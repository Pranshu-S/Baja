```python
from IPython.display import Image
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

## Graph 1 - Camber vs Rack displacement

![Graph1](Graphs/Graph1.jpeg)


```python
x = [0.0,30.0]
y = [2.7, -1.4]
```

```python
f = interp1d(x, y)
xnew = np.linspace(0, x[1], num=41, endpoint=True)
plt.plot(x, y, 'o', xnew, f(xnew))
M = (y[1] - y[0])/(x[1] - x[0])
print("Gradient = " + str(M))
```
