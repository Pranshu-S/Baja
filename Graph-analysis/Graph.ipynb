```python
from IPython.display import Image
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

## Graph 1 - Camber vs Rack displacement

![Graph1](Graphs/Graph1.jpeg)

```python
x = [-30.0, -20.0, -10.0, 0.0, 10.0, 20.0, 30.0]
y = [5.8, 4.6, 3.75, 2.7, 1.5, 0.2, -1.4]
```

```python
tck = interpolate.splrep(x, y, s=0)
test_x = np.linspace(-30,30, num=61, endpoint=True)
result_y = interpolate.splev(test_x, tck, der=0)
plt.xlabel('Length')
plt.ylabel('Angle')
plt.axis([-30, 30, -1.5, 6])
plt.grid()
plt.plot(x, y, 'x', test_x, result_y)
```

```python
yder = interpolate.splev(test_x, tck, der=1)
plt.figure()
plt.grid()
plt.plot(test_x, yder)
```
