```python
from IPython.display import Image
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
def getvals(x,y):
    print(x,y)
    f = interp1d(x, y)
    xnew = np.linspace(x[0], x[1], num=41, endpoint=True)
    plt.grid()
    plt.plot(x, y, 'o', xnew, f(xnew))
    M = (y[1] - y[0])/(x[1] - x[0])
    print("Gradient = " + str(M))
```

```python
x = [-100, -80, -60, -40, -20, 0, 20, 40, 60, 80, 100]
y = [-0.9471, -0.6632, -0.4351, -0.2529, -0.1096, 0, 0.0796, 0.1319, 0.1590, 0.1621, 0.1420]
```

```python
tck = interpolate.splrep(x, y, s=0)
test_x = np.linspace(x[0],x[10], num=61, endpoint=True)
result_y = interpolate.splev(test_x, tck, der=0)
plt.xlabel('Length')
plt.ylabel('Angle')
plt.axis([x[0]-10,x[10]+10,y[0]-0.1,y[10]+0.1,])
plt.grid()
plt.plot(x, y, 'x', test_x, result_y)
```

```python
yder = interpolate.splev(test_x, tck, der=1)
plt.figure()
plt.grid()
plt.plot(test_x, yder)
```

```python
min_max_x = [min(x),max(x)]
min_max_y = [min(y),max(y)]
```

```python
getvals(min_max_x,min_max_y)
```

## Rebound

```python
x = [0, 20, 40, 60, 80, 100]
y = [0, 0.0796, 0.1319, 0.1590, 0.1621, 0.1420]
```

```python
tck = interpolate.splrep(x, y, s=0)
test_x = np.linspace(x[0],x[5], num=61, endpoint=True)
result_y = interpolate.splev(test_x, tck, der=0)
plt.xlabel('Length')
plt.ylabel('Angle')
plt.axis([x[0]-10,x[5]+10,y[0]-0.1,y[5]+0.1,])
plt.grid()
plt.plot(x, y, 'x', test_x, result_y)
```

```python
yder = interpolate.splev(test_x, tck, der=1)
plt.figure()
plt.grid()
plt.plot(test_x, yder)
```

```python
# Taking minimum and maximum values

min_max_x = [min(x),max(x)]
min_max_y = [min(y),max(y)]
```

```python
getvals(min_max_x,min_max_y)
```

```python
# Taking endpoints

min_max_x = [x[0], x[5]]
min_max_y = [y[0],y[5]]
```

```python
getvals(min_max_x,min_max_y)
```
